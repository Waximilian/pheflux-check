```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Problema 1

def f(t):
    return 20*np.sin(np.pi*t)+150-150
def dfdx(t):
    return 20*np.cos(np.pi*t)*np.pi

def NewtonRaphson(x0):
    
    max_iter=1000
    tol=1e-6
    
    xn_1 = x0
    for n in range(1,max_iter):
        xn = xn_1 - f(xn_1)/dfdx(xn_1)
        if abs( f(xn)) < tol: break
        xn_1 = xn
    
    return(xn)

t=np.linspace(0,10,200)
y=f(t)
plt.plot(t,y)
plt.axvline(NewtonRaphson(3),color="red")
plt.grid()
```

```python
# Problema 2

def f(a,b,t):
    return 0.011825*(t - a)**2 + 1.356975*(t - a) + b - 900

def Secante(xa,xb):
    max_iter=1000
    tol=1e-6
    
    for n in range(2,max_iter):
        xc = xb - f(a,b,xb)/( (f(a,b,xb)-f(a,b,xa))/(xb-xa) )
        if abs( f(a,b,xc)) < tol: break
        xa = xb
        xb = xc
    
    return(xc)

a =  1995.7
b =  341.3

t=np.linspace(1960,2400,200)
y=f(a,b,t)
plt.plot(t,y)
plt.axvline(Secante(2200,2201),color="red")
plt.grid()
```

```python
def midPoint(t0,x0,tf,h):
    n=(int)((tf-t0)/h)
    ts,xs= [],[]
    t,x=t0,x0
    for i in range(1,n+1):
        t = t + h
        x = x + h*f(t+h/2,x+h/2*f(t,x))
        xs.append(x)
        ts.append(t)
    return ts,xs

def f(t,x):
    return 0.1*x*t**3
    
def If(t,x=1):
    return np.exp(0.1/4*t**4)

tf=1
h=0.001
t0,x0 = 0,1

ts=np.linspace(0,tf,100)
plt.plot(ts,If(np.array(ts)),color="blue")

ts,xs = midPoint(t0,x0,tf,h)
plt.plot(ts,xs,color="orange")
```
